```python
# ===================================
# 🎙️ LOCUTOR Y MICROFONO IA
# ===================================

print("===================================")
print("      LOCUTOR Y MICROFONO IA")
print("===================================")

while True:

    print("\n=========== MENU ===========")
    print("1. Revisar voz")
    print("2. Revisar microfono")
    print("3. Analizar locutor")
    print("4. Salir")

    opcion = int(input("Seleccione una opcion: "))

    # -----------------------------------
    # OPCION 1
    # -----------------------------------

    if opcion == 1:

        nombre = input("Ingrese el nombre del locutor: ")

        tono = input("Ingrese el tono de voz (grave/agudo): ")

        volumen = int(input("Ingrese nivel de volumen: "))

        print("\nANALIZANDO VOZ...\n")

        if tono == "grave" and volumen >= 8:

            print("Voz fuerte y profesional.")
            print("Ideal para radio en vivo.")

        elif tono == "agudo":

            print("Voz juvenil detectada.")
            print("Ideal para entretenimiento.")

        else:

            print("Voz estable.")
            print("Buen rendimiento de locucion.")

    # -----------------------------------
    # OPCION 2
    # -----------------------------------

    elif opcion == 2:

        microfono = input("¿El microfono funciona? (si/no): ")

        ruido = input("¿Hay ruido externo? (si/no): ")

        print("\nREVISANDO MICROFONO...\n")

        if microfono == "no":

            print("Microfono dañado.")
            print("Revisar conexiones.")

        elif ruido == "si":

            print("Ruido detectado.")
            print("Reducir sonido externo.")

        else:

            print("Microfono funcionando correctamente.")

    # -----------------------------------
    # OPCION 3
    # -----------------------------------

    elif opcion == 3:

        experiencia = int(input("Ingrese años de experiencia: "))

        horas = int(input("Ingrese horas de transmision: "))

        print("\nANALISIS DEL LOCUTOR...\n")

        if horas >= 10:

            print("Locutor con transmision extensa.")
            print("Se recomienda descanso.")

        elif experiencia >= 5:

            print("Locutor con experiencia.")
            print("Excelente desempeño.")

        else:

            print("Locutor principiante.")
            print("Seguir practicando.")

    # -----------------------------------
    # OPCION 4
    # -----------------------------------

    elif opcion == 4:

        print("\n===================================")
        print("      PROGRAMA FINALIZADO")
        print("===================================")

        print("Alumno: Marco Morquecho")
        print("Especialidad: Inteligencia Artificial")
        print("Docente: Eliseo Nava")
        print("CBTis 128")

        print("\nGracias por utilizar LOCUTOR Y MICROFONO IA")

        break

    # -----------------------------------
    # OPCION INVALIDA
    # -----------------------------------

    else:

        print("Opcion no valida")
```
