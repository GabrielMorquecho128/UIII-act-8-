```python
# ===================================
# ⚓ MARINERO Y BRUJULA IA
# ===================================

print("===================================")
print("      MARINERO Y BRUJULA IA")
print("===================================")

while True:

    print("\n=========== MENU ===========")
    print("1. Revisar ruta")
    print("2. Revisar brujula")
    print("3. Analizar marinero")
    print("4. Salir")

    opcion = int(input("Seleccione una opcion: "))

    # -----------------------------------
    # OPCION 1
    # -----------------------------------

    if opcion == 1:

        nombre = input("Ingrese el nombre del marinero: ")

        clima = input("Ingrese el clima (soleado/tormenta): ")

        olas = int(input("Ingrese nivel de olas: "))

        print("\nANALIZANDO RUTA...\n")

        if clima == "tormenta" and olas >= 8:

            print("Ruta peligrosa detectada.")
            print("Se recomienda regresar al puerto.")

        elif clima == "soleado":

            print("Ruta segura.")
            print("Navegacion estable.")

        else:

            print("Ruta moderada.")
            print("Continuar con precaucion.")

    # -----------------------------------
    # OPCION 2
    # -----------------------------------

    elif opcion == 2:

        brujula = input("¿La brujula funciona? (si/no): ")

        direccion = input("¿La direccion es correcta? (si/no): ")

        print("\nREVISANDO BRUJULA...\n")

        if brujula == "no":

            print("Brujula dañada.")
            print("Revisar sistema de navegacion.")

        elif direccion == "no":

            print("Direccion incorrecta.")
            print("Ajustar coordenadas.")

        else:

            print("Brujula funcionando correctamente.")

    # -----------------------------------
    # OPCION 3
    # -----------------------------------

    elif opcion == 3:

        experiencia = int(input("Ingrese años de experiencia: "))

        horas = int(input("Ingrese horas navegando: "))

        print("\nANALISIS DEL MARINERO...\n")

        if horas >= 12:

            print("Marinero con jornada extensa.")
            print("Se recomienda descanso.")

        elif experiencia >= 5:

            print("Marinero con experiencia.")
            print("Excelente navegacion.")

        else:

            print("Marinero principiante.")
            print("Seguir practicando.")

    # -----------------------------------
    # OPCION 4
    # -----------------------------------

    elif opcion == 4:

        print("\n===================================")
        print("      PROGRAMA FINALIZADO")
        print("===================================")

        print("Alumno: Marco Morquecho")
        print("Especialidad: Inteligencia Artificial")
        print("Materia: Elabora Proyectos de Programacion Logica")
        print("Docente: Eliseo Nava")
        print("CBTis 128")

        print("\nGracias por utilizar MARINERO Y BRUJULA IA")

        break

    # -----------------------------------
    # OPCION INVALIDA
    # -----------------------------------

    else:

        print("Opcion no valida")
```
